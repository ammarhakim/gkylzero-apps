```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the data from csv files
dir_ne = './Electron_density/'
file_ne_high = 'ne_background_probe_High_dens_33341.csv'
file_ne_low = 'ne_background_probe_Low_dens_33341.csv'
file_ne_mid = 'ne_background_probe_Middle_dens_33341.csv'

dir_Te = './Electron_temperature/'
file_Te_high = 'Te_background_probe_High_dens_33341.csv'
file_Te_low = 'Te_background_probe_Low_dens_33341.csv'
file_Te_mid = 'Te_background_probe_Middle_dens_33341.csv'

dir_qW = './Heat_flux/'
file_qW_high = 'qW_background_probe_High_dens_33341.csv'
file_qW_low = 'qW_background_probe_Low_dens_33341.csv'
file_qW_mid = 'qW_background_probe_Middle_dens_33341.csv'

df_ne_high = pd.read_csv(dir_ne+file_ne_high)
df_ne_low = pd.read_csv(dir_ne+file_ne_low)
df_ne_mid = pd.read_csv(dir_ne+file_ne_mid)

df_Te_high = pd.read_csv(dir_Te+file_Te_high)
df_Te_low = pd.read_csv(dir_Te+file_Te_low)
df_Te_mid = pd.read_csv(dir_Te+file_Te_mid)

df_qW_high = pd.read_csv(dir_qW+file_qW_high)
df_qW_low = pd.read_csv(dir_qW+file_qW_low)
df_qW_mid = pd.read_csv(dir_qW+file_qW_mid)


print(df_ne_high.head())
print(df_Te_low.head())
print(df_qW_high.head())
```

```python
# Access the column at index 0 (the first column)
RRsep_ne_high = df_ne_high.iloc[:, 0]
ne_high = df_ne_high.iloc[:, 1]
RRsep_ne_low = df_ne_low.iloc[:, 0]
ne_low = df_ne_low.iloc[:, 1]
RRsep_ne_mid = df_ne_mid.iloc[:, 0]
ne_mid = df_ne_mid.iloc[:, 1]

RRsep_Te_high = df_Te_high.iloc[:, 0]
Te_high = df_Te_high.iloc[:, 1]
RRsep_Te_low = df_Te_low.iloc[:, 0]
Te_low = df_Te_low.iloc[:, 1]
RRsep_Te_mid = df_Te_mid.iloc[:, 0]
Te_mid = df_Te_mid.iloc[:, 1]

RRsep_qW_high = df_qW_high.iloc[:, 0]
qW_high = df_qW_high.iloc[:, 1]
RRsep_qW_low = df_qW_low.iloc[:, 0]
qW_low = df_qW_low.iloc[:, 1]
RRsep_qW_mid = df_qW_mid.iloc[:, 0]
qW_mid = df_qW_mid.iloc[:, 1]

# Remove the NaN values
RRsep_ne_high = RRsep_ne_high[~np.isnan(ne_high)]
ne_high = ne_high[~np.isnan(ne_high)]
RRsep_ne_low = RRsep_ne_low[~np.isnan(ne_low)]
ne_low = ne_low[~np.isnan(ne_low)]
RRsep_ne_mid = RRsep_ne_mid[~np.isnan(ne_mid)]
ne_mid = ne_mid[~np.isnan(ne_mid)]
print(np.shape(RRsep_ne_high), np.shape(ne_high), np.shape(RRsep_ne_high)[0]-np.shape(ne_high)[0])
print(np.shape(RRsep_ne_low), np.shape(ne_low), np.shape(RRsep_ne_low)[0]-np.shape(ne_low)[0])
print(np.shape(RRsep_ne_mid), np.shape(ne_mid), np.shape(RRsep_ne_mid)[0]-np.shape(ne_mid)[0])

RRsep_Te_high = RRsep_Te_high[~np.isnan(Te_high)]
Te_high = Te_high[~np.isnan(Te_high)]
RRsep_Te_low = RRsep_Te_low[~np.isnan(Te_low)]
Te_low = Te_low[~np.isnan(Te_low)]
RRsep_Te_mid = RRsep_Te_mid[~np.isnan(Te_mid)]
Te_mid = Te_mid[~np.isnan(Te_mid)]
print(np.shape(RRsep_Te_high), np.shape(Te_high), np.shape(RRsep_Te_high)[0]-np.shape(Te_high)[0])
print(np.shape(RRsep_Te_low), np.shape(Te_low), np.shape(RRsep_Te_low)[0]-np.shape(Te_low)[0])
print(np.shape(RRsep_Te_mid), np.shape(Te_mid), np.shape(RRsep_Te_mid)[0]-np.shape(Te_mid)[0])

RRsep_qW_high = RRsep_qW_high[~np.isnan(qW_high)]
qW_high = qW_high[~np.isnan(qW_high)]
RRsep_qW_low = RRsep_qW_low[~np.isnan(qW_low)]
qW_low = qW_low[~np.isnan(qW_low)]
RRsep_qW_mid = RRsep_qW_mid[~np.isnan(qW_mid)]
qW_mid = qW_mid[~np.isnan(qW_mid)]
print(np.shape(RRsep_qW_high), np.shape(qW_high), np.shape(RRsep_qW_high)[0]-np.shape(qW_high)[0])
print(np.shape(RRsep_qW_low), np.shape(qW_low), np.shape(RRsep_qW_low)[0]-np.shape(qW_low)[0])
print(np.shape(RRsep_qW_mid), np.shape(qW_mid), np.shape(RRsep_qW_mid)[0]-np.shape(qW_mid)[0])
```

```python
# Plot the ne, Te, and qW profiles
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, figsize=[4.8, 6.4], sharex=True)

xlim = [0.98, 1.04]
ylim_ne = [0.0, 6.4]
ylim_Te = [0.0, 24.0]
ylim_qW = [0.005, 2.0]

ax1.plot(RRsep_ne_low, ne_low, color='red', label='low')
ax1.plot(RRsep_ne_mid, ne_mid, color='black', label='mid')
ax1.plot(RRsep_ne_high, ne_high, color='blue', label='high')
ax1.vlines(1.0, ylim_ne[0], ylim_ne[1], color='k', linestyle='--')
ax1.set_xlim(xlim[0], xlim[1])
ax1.set_ylim(ylim_ne[0], ylim_ne[1])
ax1.set_ylabel(r'$n_e~\rm{(m^{-3})}$')
ax1.legend()

ax2.plot(RRsep_Te_low, Te_low, color='red', label='low')
ax2.plot(RRsep_Te_mid, Te_mid, color='black', label='mid')
ax2.plot(RRsep_Te_high, Te_high, color='blue', label='high')
ax2.vlines(1.0, ylim_Te[0], ylim_Te[1], color='k', linestyle='--')
ax2.set_xlim(xlim[0], xlim[1])
ax2.set_ylim(ylim_Te[0], ylim_Te[1])
ax2.set_ylabel(r'$T_e~\rm{(eV)}$')
ax2.legend()

ax3.semilogy(RRsep_qW_low, qW_low, color='red', label='low')
ax3.semilogy(RRsep_qW_mid, qW_mid, color='black', label='mid')
ax3.semilogy(RRsep_qW_high, qW_high, color='blue', label='high')
ax3.vlines(1.0, ylim_qW[0], ylim_qW[1], color='k', linestyle='--')
ax3.set_xlim(xlim[0], xlim[1])
ax3.set_ylim(ylim_qW[0], ylim_qW[1])
ax3.set_xlabel(r'$\rho_p$')
ax3.set_ylabel(r'$q_w~\rm{(MW/m^2)}$')
ax3.legend()

plt.tight_layout()
```
